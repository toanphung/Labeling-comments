```python
import os
```

```python
from pyvi import ViTokenizer, ViPosTagger


pos_sample = {"A": "Tính từ", "E": "Giới từ", "I": "Thán từ", "L": "Từ xác định", "M": "Chữ số", "N": "Danh từ chung",
            "Nc": "Danh từ phân loại", "Ny": "Danh từ viết tắt", "Np": "Danh từ riêng", "Nu": "Danh từ đơn vị",
            "P": "Đại từ", "R": "Trạng từ", "V": "Động từ", "X": "Không rõ", "F": "Dấu câu", "C": "Liên từ",
            "S": "Liên từ phụ", "T": "Từ phụ trợ"}

def get_pos_tagging(string):
    return ViPosTagger.postagging(ViTokenizer.tokenize(u""+string))[1]
```

```python
import pickle
# load list of stop words
pickle_in = open("/".join(os.getcwd().split('/')[:-1])+ "/stop_words_vietnamese.pickle", "rb")
list_stop_word = pickle.load(pickle_in)
# load abbreviation words
pickle_in = open("/".join(os.getcwd().split('/')[:-1])+ "/abbreviation_words_vietnamese.pickle", "rb")
abbreviation = pickle.load(pickle_in)
```

```python
def analyze_data(string, dict_):
    global pos_dict
    for ch in string.split():
        if ch in list_stop_word or (not ch.isalnum() and ch.find("_") == -1):
            continue 
        if ch in abbreviation:
            ch = abbreviation[ch]
        if ch.lower() not in dict_:
            dict_[ch.lower()] = 0
        dict_[ch.lower()] += 1
        pos = pos_sample[get_pos_tagging(ch)[0]]
        if pos not in pos_dict:
            pos_dict[pos] = 0
        pos_dict[pos] += 1
    return dict_
```

```python
client_type_dict = dict()
rating_dict = dict()
title_dict = dict()
content_dict = dict()
pos_dict = dict()
length_sentence_title = []
length_sentence_content = []
count = 2


with open('./comments_lazada_data.csv', 'r') as f:
    f.readline() # ignore the first line
    for line in f.readlines():
        # ------------ Preprocessing -----------------------------------
        # There is double quotes mixing with "," in the data, so I firstly process it
        temp = line
        inds =  []
        occurence = temp.find('"')
        off_set = 0
        while occurence != -1:
            temp = temp[:occurence] + temp[occurence+1:]
            inds.append(occurence + off_set)
            occurence = temp.find('"')
            off_set += 1
        list_ = []
        if len(inds):
            list_.extend(line[:inds[0]].strip().split(","))
            for i in range(0, len(inds), 2):
                list_.append(line[inds[i]+1:inds[i+1]])
            if len(line[inds[-1]:]):
                list_.append(line[inds[-1]+2:])
        else:
            list_ = line.split(",")
        # ---------------------------------------------------------------
        # ---------- Feed data to do statistic --------------------------
        list_ = [l.strip() for l in list_ if len(l) > 0]
        print(count, list_)
        client_type, rating, title, *content = list_
        content = " ".join(content)
        if len(title) > 0: length_sentence_title.append(len(title))
        if len(content) > 0: length_sentence_content.append(len(content))
        if client_type not in client_type_dict:
            client_type_dict[client_type] = 0
        client_type_dict[client_type] += 1
        if rating not in rating_dict:
            rating_dict[rating] = 0
        rating_dict[rating] += 1
        title_dict = analyze_data(title, title_dict)
        content_dict = analyze_data(content, content_dict)
        # _---------------------------------------------------------------
        count += 1
```

```python
popular_title_words = sorted(title_dict.items(), key=lambda x: x[1], reverse=True)[:50]
popular_content_words = sorted(content_dict.items(), key=lambda x: x[1], reverse=True)[:50]
```

```python
print("Top 50 popular words in title:")
for k, v in popular_title_words:
    print(k.ljust(10), pos_sample[get_pos_tagging(k)[0]].ljust(15), v)
```

```python
print("Top 50 popular words in content:")
for k, v in popular_content_words:
    print(k.ljust(10), pos_sample[get_pos_tagging(k)[0]].ljust(15), v)
```

```python
print("Number of rating for each: ")
for k, v in sorted(rating_dict.items()):
    print(k+" sao:", v)
```

```python
print("Number of devices for each client type: ")
max_length = len(max(client_type_dict.items(), key=lambda x: len(x[0]))[0])
for k, v in sorted(client_type_dict.items()):
    print(k.ljust(max_length), v)
```

```python
length_sentence_title.sort()
length_sentence_content.sort()
n = len(length_sentence_title)
print("Analyze length of the title column:")
print("Max length: ", length_sentence_title[-1])
print("Min length: ", length_sentence_title[0])
print("Average: ", sum(length_sentence_title)/n)
print("Median: ", (length_sentence_title[n//2] if n & 1 else 
                    (length_sentence_title[n//2]+length_sentence_title[n//2+1])/2))
```
